```python
# 3번 움직여
# 비싼곳으로 이동해
# 총합을 계산해
# 각 시작 지점에서 맥스 값

# n, m

def go(i, j, cost=0, step=0, pre=None, straight=False):
  
  def valid(_i, _j, pre):
    if (_i, _j) != pre and (0 <= _i < n ) and (0<=_j < m):
        return True
    return False
  
  def isStraight(pre, curr, future):
    if pre[0] == curr[0] == future[0] or pre[1] == curr[1] == future[1]:
      return True
    
    return False
  
  # if step == 2:
  #   print(f"{step} -- {i, j}, {pre}")
  
  case = [ (i-1, j), (i+1, j), (i, j-1), (i, j+1) ]
  
  if straight and pre[0] == i:
    case.extend([ (i-1, j-1), (i+1, j-1)])
  elif straight and pre[1] == j:
    case.extend([ (i-1, j-1), (i-1, j+1) ])
  
  print(f"{i,j}--{case}")
  large = -1
  future = i, j 
  for n_i, n_j in case:
    if not valid(n_i, n_j, pre): 
      continue
  
    if mat[n_i][n_j] > large:
      # print("hi")
      large = mat[n_i][n_j] 
      future = n_i, n_j
  
    
  # print(f"{step} -- {pre} -> {i, j} -> {future} (isStraight:{straight})")

  if step == 1 :
    straight = isStraight(pre, (i,j), future)

  if step < 2:
    acc_cost = cost + mat[future[0]][future[1]]
    # print(acc_cost)
    cost = go(future[0], future[1], cost=acc_cost, step=step+1, pre=(i,j), straight=straight)
  
  else:
    acc_cost = cost + mat[i][j]
    # print(f"end -- {i,j, mat[i][j]} -- {acc_cost}")
    
    return(acc_cost)
  
  return cost

val = go(3, 0)

print(val)

      
```

```python
# n, m = 5, 5
# inputs = [
#   "1 2 3 4 5",
#   "5 4 3 2 1",
#   "2 3 4 5 6",
#   "6 5 4 3 2",
#   "1 2 1 2 1"
# ]

# mat = [list(map(int, inputs[i].split())) for i in range(n)]

import sys
input = sys.stdin.readline
n, m = map(int, input().split())
# mat = [list(map(int, sys.stdin.readline().split())) for _ in range(n)]

mat = [[] for _ in range(n)]
for i in range(n):
    mat[i] = list(map(int, input().split()))
    
dx = [0, 0, -1, 1]
dy = [-1, 1, 0, 0]

visited = [[False for _ in range(m)] for _ in range(n)]

max_cost = -1

# def isMovable(x, y):
#     global n, m
    
#     return (0<=x < n and 0<= y < m)

def gogo(x, y, nblocks, cost):
    global n, m, mat, dx, dy, max_cost
    
    if nblocks == 4:
        max_cost = max(max_cost, cost)
        return

    for i in range(4):
        nx, ny = x+dx[i], y+dy[i]

        if nx < 0 or ny < 0 or nx >= n or ny >= m:
            continue
        
        if not visited[nx][ny]:
            visited[nx][ny] = True
            gogo(nx, ny, nblocks+1, cost + mat[nx][ny])
            visited[nx][ny] = False


def main():
    global n, m, mat, max_cost

    for i in range(n):
        for j in range(m):
            visited[i][j] = True
            gogo(i, j, 1, mat[i][j])
            visited[i][j] = False

            if i < n - 2 and j < m - 1:
                max_cost = max(max_cost, mat[i][j] + mat[i+1][j] + mat[i+1][j+1] + mat[i+2][j])
            # ㅜ
            if i < n - 1 and j < m - 2:
                max_cost = max(max_cost, mat[i][j] + mat[i][j+1] + mat[i][j+2] + mat[i+1][j+1])
            # ㅗ
            if i >= 1 and j < m - 2:
                max_cost = max(max_cost, mat[i][j] + mat[i-1][j+1] + mat[i][j+1] + mat[i][j+2])
            # ㅓ
            if 1 <= i < n - 1 and j < m - 1:
                max_cost = max(max_cost, mat[i][j] + mat[i-1][j+1] + mat[i][j+1] + mat[i+1][j+1])

main()
print(max_cost)
```
